```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

# IMPORTING THE DATASET

```python
dataset=pd.read_csv("E:/credit card fraud detection/creditcard/creditcard.csv")
```

```python
## separating the data fro analysis
true=dataset[dataset.Class==0]
false=dataset[dataset.Class==1]
```

# UNDERSAMPLING FOR IMBALANCED DATASET

```python
#undersampling
true_sample=true.sample(n=492)
```

```python
##concatenate the new treu dataset and the false dataset

new_dataset = pd.concat([true_sample,false],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset["Class"].value_counts()
```

```python
new_dataset.groupby("Class").mean()
```

```python
## splitting the data into features and targets
x= new_dataset.drop(columns='Class',axis=1)
y= new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

# TEST TRAIN SPLIT

```python
##splitting the data into test and train sets
X_train,X_test,Y_train,Y_test = train_test_split(x,y, test_size=0.2, stratify=y , random_state=2)
                                                 
```

```python
print(x.shape, X_train.shape, X_test.shape)
```

# LOGISTIC REGRESSION

```python
##Logistic regression
model = LogisticRegression()
model.fit(X_train,Y_train)
X_train_prediction=model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print("training data accuracy :", training_data_accuracy)
```

```python
## testing data acciracy
X_test_prediction=model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print("testing data accuracy :", testing_data_accuracy)
```
